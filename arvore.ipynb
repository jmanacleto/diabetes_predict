## Importação de Bibliotecas

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

## Carregamento dos Dados

```python
data = pd.read_csv('https://raw.githubusercontent.com/jmanacleto/diabetes_predict/main/diabetes.csv')
```

```python
data.head()
```

## Separação das Features e do Target

```python
# Dividir as features e o target
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

## Divisão dos Dados em Conjuntos de Treinamento e Teste

```python
# Dividir o dataset em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Criação e Treinamento do Modelo de Árvore de Decisão

Function

```python
# Criar o modelo de árvore de decisão
clf = DecisionTreeClassifier(random_state=42)
```

# Máquina de vetores de suporte

Não precisei instanciar nenhum tipo de modelo de arvore

entropia ou índice de Gini

```python
# Treinar o modelo
clf.fit(X_train, y_train)
```

## Previsão e Avaliação do Modelo

```python
# Fazer previsões
y_pred = clf.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy:.2f}')

print('Relatório de Classificação:')
print(classification_report(y_test, y_pred))

print('Matriz de Confusão:')
print(confusion_matrix(y_test, y_pred))
```

## A acurácia
é a proporção de previsões corretas em relação ao total de previsões. Aqui, o modelo tem uma acurácia de 70%, o que significa que ele classifica corretamente 70% dos casos.

## **Classe 0 (No Diabetes)**
**(Precisão): 0.81**
Isto significa que 81% das previsões de "No Diabetes" estavam corretas.

**Recall (Sensibilidade): 0.71**
Isto significa que 71% dos casos reais de "No Diabetes" foram corretamente identificados pelo modelo.

**F1-score: 0.76%**
O F1-score é a média harmônica da precisão e do recall, proporcionando uma medida equilibrada da precisão do modelo.

## **Classe 1 (Diabetes)**
**Precision (Precisão): 0.56**
Isto significa que 56% das previsões de "Diabetes" estavam corretas.

**Recall (Sensibilidade): 0.69**
Isto significa que 69% dos casos reais de "Diabetes" foram corretamente identificados pelo modelo.

**F1-score: 0.61**
O F1-score para a classe "Diabetes" é menor, indicando que o modelo é menos eficiente para essa classe.


## A matriz de confusão mostra a distribuição das previsões:

**True Positives (TP): 107**
Casos de "No Diabetes" corretamente classificados como "No Diabetes".

**False Positives (FP): 44**
Casos de "Diabetes" incorretamente classificados como "No Diabetes".

**False Negatives (FN): 25**
Casos de "No Diabetes" incorretamente classificados como "Diabetes".

**True Negatives (TN): 55**
Casos de "Diabetes" corretamente classificados como "Diabetes".

## Visualização da Árvore de Decisão

```python
# Visualizar a árvore de decisão #!!!!
plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=X.columns, class_names=['No Diabetes', 'Diabetes'], filled=True)
plt.show()
```

**Nós Internos:** Cada nó interno representa um teste em um atributo, como níveis de glicose, pressão arterial, etc.

**Ramos:** Cada ramo indica o resultado do teste. Por exemplo, se a glicose é maior que um certo valor, siga para a esquerda; caso contrário, para a direita.

**Nós Folha:** Cada nó folha representa uma classificação final, que neste caso seria ‘diabetes’ ou ‘não diabetes’.

**Caminhos:** Os caminhos da raiz até as folhas representam as regras de classificação baseadas nos atributos testados.

## Modelo **SVM**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

## Separação das Features e do Target

```python
# Dividir as features e o target
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

## Divisão dos Dados em Conjuntos de Treinamento e Teste

```python
# Dividir o dataset em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Criação e Treinamento do Modelo SVM

```python
# Criar o modelo SVM
svm = SVC(kernel='linear', random_state=42)
```

 Classificação e regressão.

```python
# Treinar o modelo
svm.fit(X_train, y_train)
```

## Previsão e Avaliação do Modelo SVM

```python
# Fazer previsões
y_pred = svm.predict(X_test)
```

```python
# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy:.2f}')

print('Relatório de Classificação:')
print(classification_report(y_test, y_pred))

print('Matriz de Confusão:')
print(confusion_matrix(y_test, y_pred))
```

## Acurácia

é a proporção de previsões corretas sobre o total de previsões. Aqui, o modelo tem uma acurácia de 74%, indicando que ele classifica corretamente 74% dos casos.

## Classe 0 (No Diabetes)

**F1-score: 0.81**
O F1-score é a média harmônica da precisão e do recall, proporcionando uma medida equilibrada da precisão do modelo.
